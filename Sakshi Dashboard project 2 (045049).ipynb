```python
import pandas as pd
import dash
import dash_core_components as dcc
import dash_html_components as html
from dash.dependencies import Input, Output
import plotly.express as px

# Load the data
filename = "covid 19.xlsx"
df = pd.read_excel(filename)

# Initialize the Dash app
app = dash.Dash(__name__)

# Define the layout of the app
app.layout = html.Div([
    html.H1("COVID-19 Dashboard"),

    dcc.Dropdown(
        id='country-dropdown',
        options=[
            {'label': country, 'value': country} for country in df['Country/Region'].unique()
        ],
        value=df['Country/Region'].unique()[0],
        multi=False,
        style={'width': '50%'}
    ),

    dcc.Dropdown(
        id='region-slicer',
        options=[
            {'label': region, 'value': region} for region in df['WHO Region'].unique()
        ],
        value=df['WHO Region'].unique(),
        multi=True,
        style={'width': '50%'}
    ),

    dcc.Graph(id='confirmed-line-chart'),
    dcc.Graph(id='deaths-bar-chart'),
    dcc.Graph(id='recovered-pie-chart'),
    dcc.Graph(id='active-scatter-plot'),

    html.Div(id='selected-country-stats')
])

# Define callback to update charts and display statistics based on dropdown selection
@app.callback(
    [Output('confirmed-line-chart', 'figure'),
     Output('deaths-bar-chart', 'figure'),
     Output('recovered-pie-chart', 'figure'),
     Output('active-scatter-plot', 'figure'),
     Output('selected-country-stats', 'children')],
    [Input('country-dropdown', 'value'),
     Input('region-slicer', 'value')]
)
def update_charts_and_stats(selected_country, selected_regions):
    filtered_df = df[df['Country/Region'] == selected_country]
    filtered_df = filtered_df[filtered_df['WHO Region'].isin(selected_regions)]

    # Line Chart - Confirmed Cases Over Time
    confirmed_line_chart = px.line(
        filtered_df,
        x=filtered_df.index,
        y='Confirmed',
        line_group='Country/Region',
        color='Country/Region',
        labels={'Confirmed': 'Confirmed Cases'},
        title='Confirmed Cases Over Time'
    )

    # Bar Chart - Deaths
    deaths_bar_chart = {
        'data': [{'x': ['Confirmed', 'Deaths', 'Recovered', 'Active'],
                  'y': [filtered_df['Confirmed'].max(), filtered_df['Deaths'].max(), filtered_df['Recovered'].max(), filtered_df['Active'].max()],
                  'type': 'bar'}],
        'layout': {'title': 'COVID-19 Statistics'}
    }

    # Pie Chart - Recovered Percentage
    recovered_pie_chart = {
        'data': [{'labels': ['Active', 'Recovered'],
                  'values': [filtered_df['Active'].max(), filtered_df['Recovered'].max()],
                  'type': 'pie'}],
        'layout': {'title': 'Active vs Recovered Percentage'}
    }

    # Scatter Plot - 1 Week Change vs 1 Week % Increase
    active_scatter_plot = px.scatter(
        filtered_df,
        x='1 week change',
        y='1 week % increase',
        color='WHO Region',
        labels={'1 week change': '1 Week Change', '1 week % increase': '1 Week % Increase'},
        title='1 Week Change vs 1 Week % Increase'
    )

    selected_country_stats = f"Selected Country: {selected_country}, Total Confirmed: {filtered_df['Confirmed'].max()}, Total Deaths: {filtered_df['Deaths'].max()}, Total Recovered: {filtered_df['Recovered'].max()}"

    return confirmed_line_chart, deaths_bar_chart, recovered_pie_chart, active_scatter_plot, selected_country_stats

# Run the app
if __name__ == '__main__':
    app.run_server(port=8050, debug=True)
```
